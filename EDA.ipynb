# Analysis
- When we look at the value of Area, the curve is not symmetric.
- They are skewed - values are clustered in a range, but many extream values(outliers) stretch the distribution toward on side.

# Why this is a problem for ML?
- Many machine learning algorithms(like linear regression, logistic regression, KNN, Neural network) are sensitive for scaling.
- Outliers can distort scaling methods(like MinMaxScaler or StandardScaler) - making most data ppear compressed into small range.
- As a result, the model may give too much importance to these rare extreme values.

# Solution - using the clip function
- clip() in pandas limits values to within a given range
- by clipping extream values, you reduce skewness in the distribution
- Scaling method works more effectively
- The model becames more robust and doesn't overfit to rare outliers
