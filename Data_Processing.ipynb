# What We're Dealing With
1. Sq. Ft/Feet	= 1,090 entries	= Easy - extract numbers
2. Sq. Meter	= 26 entries	= Convert to sq. ft (×10.764)
3. Ropani =	20 entries	= Complex format
4. Aana =	23 entries	= Convert to sq. ft
5. Acres = 1 entry	= Convert to sq. ft
6. Negative	= 22 entries	= Set to NaN (errors)
7. Zero =	34 entries	= Set to NaN (not meaningful)

```python
#create clean column for area sl per fit
df['BUILDUP_AREA_SQFT'] = None
```

```python
# Mark rows with negative or zero as invalid
invalid_mask = df['BUILDUP AREA'].astype(str).str.contains('^-|^0 ', na=False)
print(f"Invalid rows (negative/zero): {invalid_mask.sum()}")
# We'll leave these as None
```

```python
# Pattern: any number followed by 'sq' (case insensitive)
sqft_mask = df['BUILDUP AREA'].str.contains('sq', na=False, case=False)

def extract_number(value):
    if pd.isna(value):
        return None
    import re
    match = re.search(r'(\d+\.?\d*)', str(value))
    if match:
        return float(match.group(1))
    return None

df.loc[sqft_mask, 'BUILDUP_AREA_SQFT'] = df.loc[sqft_mask, 'BUILDUP AREA'].apply(extract_number)
```

```python
# Convert sq. meter to sq. ft (× 10.764)
meter_mask = df['BUILDUP AREA'].str.contains('meter', na=False, case=False)

def convert_meter_to_sqft(value):
    if pd.isna(value):
        return None
    import re
    match = re.search(r'(\d+\.?\d*)', str(value))
    if match:
        return float(match.group(1)) * 10.764
    return None

df.loc[meter_mask, 'BUILDUP_AREA_SQFT'] = df.loc[meter_mask, 'BUILDUP AREA'].apply(convert_meter_to_sqft)
```

```python
# 1 aana = 342.25 sq. ft
aana_mask = df['BUILDUP AREA'].str.contains('aana', na=False, case=False)

def convert_aana_to_sqft(value):
    if pd.isna(value):
        return None
    import re
    match = re.search(r'(\d+\.?\d*)', str(value))
    if match:
        return float(match.group(1)) * 342.25
    return None

df.loc[aana_mask, 'BUILDUP_AREA_SQFT'] = df.loc[aana_mask, 'BUILDUP AREA'].apply(convert_aana_to_sqft)
```

```python
# 1 acre = 43,560 sq. ft
acres_mask = df['BUILDUP AREA'].str.contains('acres', na=False, case=False)

def convert_acres_to_sqft(value):
    if pd.isna(value):
        return None
    import re
    match = re.search(r'(\d+\.?\d*)', str(value))
    if match:
        return float(match.group(1)) * 43560
    return None

df.loc[acres_mask, 'BUILDUP_AREA_SQFT'] = df.loc[acres_mask, 'BUILDUP AREA'].apply(convert_acres_to_sqft)
```

```python
# 1 Ropani = 64 aana = 5476 sq. ft (approx)
# Actually, let's research exact conversion later

ropani_mask = df['BUILDUP AREA'].str.contains('ropani', na=False, case=False)

def convert_ropani_to_sqft(value):
    if pd.isna(value):
        return None
    # Format: Ropani-Aana-Paisa-Daam
    # Example: "0-3-2-0" = 0 Ropani, 3 Aana, 2 Paisa, 0 Daam
    import re
    match = re.search(r'(\d+)-(\d+)-(\d+)-(\d+)', str(value))
    if match:
        ropani = float(match.group(1))
        aana = float(match.group(2))
        paisa = float(match.group(3))
        daam = float(match.group(4))
        
        # 1 Ropani = 64 Aana
        # 1 Aana = 4 Paisa
        # 1 Paisa = 4 Daam
        total_aana = (ropani * 64) + aana + (paisa / 4) + (daam / 16)
        return total_aana * 342.25  # Convert to sq. ft
    return None

df.loc[ropani_mask, 'BUILDUP_AREA_SQFT'] = df.loc[ropani_mask, 'BUILDUP AREA'].apply(convert_ropani_to_sqft)
```

```python
#Handle Multiple Values (2600+1600 Sq. Feet)
# For the one with "+", take the sum or first value?
multi_mask = df['BUILDUP AREA'].str.contains(r'\+', na=False)

def handle_multiple(value):
    if pd.isna(value):
        return None
    import re
    numbers = re.findall(r'(\d+\.?\d*)', str(value))
    if numbers:
        return sum(float(n) for n in numbers)  # Sum both
    return None

df.loc[multi_mask, 'BUILDUP_AREA_SQFT'] = df.loc[multi_mask, 'BUILDUP AREA'].apply(handle_multiple)
```

```python
# Remove negative and zero after conversion
df.loc[df['BUILDUP_AREA_SQFT'] <= 0, 'BUILDUP_AREA_SQFT'] = None
```

```python
def extract_single_feet(value):
    if pd.isna(value):
        return None
    import re
    # Look for number followed by Feet/FT (case insensitive)
    match = re.search(r'(\d+\.?\d*)\s*(?:Feet|FT|feet|ft)', str(value), re.IGNORECASE)
    if match:
        return float(match.group(1))
    return None

simple_mask = df['ROAD ACCESS'].str.contains(r'\d+\s*(?:Feet|FT|feet|ft)', na=False, case=False)
df.loc[simple_mask, 'ROAD_ACCESS_FEET'] = df.loc[simple_mask, 'ROAD ACCESS'].apply(extract_single_feet)
```

```python
# Convert meters to feet (× 3.28084)
def convert_meter_to_feet(value):
    if pd.isna(value):
        return None
    import re
    match = re.search(r'(\d+\.?\d*)\s*(?:Meter|meteRs|M)', str(value), re.IGNORECASE)
    if match:
        return float(match.group(1)) * 3.28084
    return None

meter_mask = df['ROAD ACCESS'].str.contains('meter', na=False, case=False)
df.loc[meter_mask, 'ROAD_ACCESS_FEET'] = df.loc[meter_mask, 'ROAD ACCESS'].apply(convert_meter_to_feet)
```

```python
# If value is just a number (like '20', '13', '10')
def extract_plain_number(value):
    if pd.isna(value):
        return None
    import re
    # Check if string is just number (no letters)
    if re.match(r'^\d+\.?\d*$', str(value).strip()):
        return float(value)
    return None

plain_mask = df['ROAD ACCESS'].astype(str).str.match(r'^\d+\.?\d*$', na=False)
df.loc[plain_mask, 'ROAD_ACCESS_FEET'] = df.loc[plain_mask, 'ROAD ACCESS'].apply(extract_plain_number)
```

```python
# For '13-20 Feet' - what to do?
# Take average (16.5) or minimum (13)

def extract_range_average(value):
    if pd.isna(value):
        return None
    import re
    # Look for pattern like '13-20'
    match = re.search(r'(\d+\.?\d*)\s*-\s*(\d+\.?\d*)', str(value))
    if match:
        low = float(match.group(1))
        high = float(match.group(2))
        return (low + high) / 2  # Average
    return None

range_mask = df['ROAD ACCESS'].str.contains(r'\d+\.?\d*\s*-\s*\d+\.?\d*', na=False)
df.loc[range_mask, 'ROAD_ACCESS_FEET'] = df.loc[range_mask, 'ROAD ACCESS'].apply(extract_range_average)
```

```python
# For '12/20 Feet' - treat same as range
slash_mask = df['ROAD ACCESS'].str.contains(r'\d+\.?\d*/\d+\.?\d*', na=False)
df.loc[slash_mask, 'ROAD_ACCESS_FEET'] = df.loc[slash_mask, 'ROAD ACCESS'].apply(extract_range_average)
```

```python
# For '12 & 22 Feet'
amp_mask = df['ROAD ACCESS'].str.contains(r'&', na=False)
df.loc[amp_mask, 'ROAD_ACCESS_FEET'] = df.loc[amp_mask, 'ROAD ACCESS'].apply(extract_range_average)
```
